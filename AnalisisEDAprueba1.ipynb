```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_validate
```

```python
X1= df[['Discount_offered', 'Weight_in_gms','Cost_of_the_Product','Customer_rating']]
Y1 = df['Reached.on.Time_Y.N']
```

```python

from sklearn.preprocessing import StandardScaler
scl = StandardScaler()
X1 = scl.fit_transform(X1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X1, Y1, random_state=10)
```

```python
print('ProporciÃ³n de etiquetas en set de entrenamiento:', y_train.sum()/y_train.count())
print('ProporciÃ³n de etiquetas en set de testeo:', y_test.sum()/y_test.count())
```

```python

for i in range(X_train.shape[1]):
    sns.distplot(x = X_train[:,i])
    sns.distplot(x = X_test[:,i])
    titulo = 'Atributo', i
    plt.title(titulo)
    plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Instanciamos el Ã¡rbol

tree = DecisionTreeClassifier(max_depth = None)
tree.fit(X_train, y_train)
```

```python
# Predecimos sobre nuestro set de entrenamiento

y_train_pred = tree.predict(X_train)

# Predecimos sobre nuestro set de test

y_test_pred = tree.predict(X_test)

# Comparamos con las etiquetas reales

print('Accuracy sobre conjunto de Train:', accuracy_score(y_train_pred,y_train))
print('Accuracy sobre conjunto de Test:', accuracy_score(y_test_pred,y_test))
```

```python

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

```python
# Predecimos sobre nuestro set de entrenamiento

y_train_pred = knn.predict(X_train)

# Predecimos sobre nuestro set de test

y_test_pred = knn.predict(X_test)

# Comparamos con las etiquetas reales
print('Accuracy sobre conjunto de Train:', accuracy_score(y_train_pred,y_train))
print('Accuracy sobre conjunto de Test:', accuracy_score(y_test_pred,y_test))
```

```python
# Importamos cross_val_score 

from sklearn.model_selection import cross_val_score
```

```python
tree = DecisionTreeClassifier()
tree_scores = cross_val_score(tree, X1, Y1, cv=5)

knn = KNeighborsClassifier(n_neighbors=1)
knn_scores = cross_val_score(knn, X1, Y1, cv=5)
```

```python
print(tree_scores)
print(knn_scores)
```

```python
print("Accuracy para Tree: %0.2f (+/- %0.2f)" % (tree_scores.mean(), tree_scores.std() * 2))
print("Accuracy para KNN: %0.2f (+/- %0.2f)" % (knn_scores.mean(), knn_scores.std() * 2))
```

```python
# Ahora probamos con cross_validate. Esta funciÃ³n brinda un poco mÃ¡s de informaciÃ³n

from sklearn.model_selection import cross_validate
```

```python
tree_scores = cross_validate(tree, X1, Y1, cv=5, scoring = ['precision','recall'])
knn_scores = cross_validate(knn, X1, Y1, cv=5, scoring = ['precision','recall'])
```
